```python
import pandas as pd
```

```python
df = pd.read_csv('combined_data_1.txt',
                encoding = 'ISO-8859-1', 
                header = None,      
                names = ["Customer_id","Rating", "Date"])
df.head()
```

```python
temp_df = df[df.Rating.isna()]
temp_df
```

```python
null_row_index =  temp_df.index.tolist()

Movie_index = 0
j = 0
for i in range(len(null_row_index)-1):
    Movie_index = Movie_index + 1        
    
    while (j != null_row_index[i+1]): 
        #for j in range (null_row_index[j],null_row_index[j+1]):
        df.loc[j,'Movie_id'] = Movie_index
        j = j+1
```

```python
df = df.dropna()
```

```python
df.sort_values(by=['Customer_id','Date'], inplace=True)
```

```python
df2 = df.groupby('Customer_id')['Movie_id'].apply(list).reset_index(name='movie_ids')
```

```python
ratings = pd.DataFrame(ratings)
```
